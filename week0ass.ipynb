```python
import pandas as pd
import numpy as np
df_meal = pd.read_csv("meal_info.csv")
df_meal.head()
```

```python
df_center = pd.read_csv("fulfilment_center_info.csv")
df_center.head()
```

```python
df_food = pd.read_csv("train.csv")
df_food.head()
```

```python
df=pd.concat([df_meal,df_center,df_food])
df.head()
```

```python
#pivot table is used to change data table and aggregate data 
#syntax index="desired tables row_name" columns=" desired tables column name"  values="desired tables values" aggfunc ="function"
table = pd.pivot_table(data=df,index='category',values='num_orders',aggfunc=np.sum)
```

```python
table.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
#bar graph
x=df['category']
y=df['num_orders']
print(x)
print(y)
```

```python
x_pos = [i for i, _ in enumerate(x)]
plt.bar(x_pos,y, color='blue')
#xticks 
#plt.xticks(rotation=70) '''Write on comment what you feel this function does'''
#rotates the graph about x-axis

#x-axis labels 
plt.xlabel('category')
#y-axis labels 
plt.ylabel('num_orders')

#plot title 
plt.title('Most popular food')

#save plot 
#savefig('C:\Users\ROSHNI\Desktop\summer ai')

#display 
plt.show()
```
